```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
X=[]
for i in range(0,7501):
      X.append([str(dataset.values[i,j])  for j in range(0,20)])
```

**Trainign Apriori Model**

```python
from apyori import apriori
rules=apriori(transactions=X,min_support=0.003,min_confidence=0.2,min_lift=3,min_length=2,max_length=2)
```

**Visualizing the results coming directly from output of apriori functions**

```python
result=list(rules)
result
```

 **Putting results well organized into pandas dataframe**

```python
def inspect(result):
  lhs=[tuple(i[2][0][0])[0] for i in result]
  rhs=[tuple(i[2][0][1])[0] for i in result]
  supports=[i[1] for i in result]
  confidences=[i[2][0][2] for i in result]
  lifts=[i[2][0][3] for i in result]
  return list(zip(lhs,rhs,supports,confidences,lifts))
resultsinDataFrame=pd.DataFrame(inspect(result),columns=['Left Hand Side','Right Hand Side','Support','Confidence','Lifts'])
```

**Displaying the results in non-sorted order**

```python
resultsinDataFrame
```

**Displaying results in sorted order**

```python
resultsinDataFrame.nlargest(n=10,columns='Lifts')
```
