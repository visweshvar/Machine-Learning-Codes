**Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Import Dataset**

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
X=dataset.iloc[:, :-1].values
Y=dataset.iloc[:,-1].values
```

**Splitting Dataset**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

**Training Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression
lrobj=LogisticRegression(random_state=0)
lrobj.fit(X_train,Y_train)
```

**Predicting New results**

```python
prediction=lrobj.predict(sc_X.transform([[30,87000]]))
print("Prediction:",prediction)
```

**Predicting Testset**

```python
y_prediction=lrobj.predict(X_test)
print(np.concatenate((y_prediction.reshape(len(y_prediction),1),Y_test.reshape(len(Y_test),1)),1))
```

**Confusion Matrix**

```python
 from sklearn.metrics import confusion_matrix,accuracy_score
 cm=confusion_matrix(Y_test,y_prediction)
 print(cm)
 accuracy_score(Y_test,y_prediction)
```

**Visualizing Trianing Set**

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc_X.inverse_transform(X_train), Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, lrobj.predict(sc_X.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('Green', 'Blue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc_X.inverse_transform(X_test), Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, lrobj.predict(sc_X.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('Green', 'Blue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
