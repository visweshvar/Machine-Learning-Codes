**Importing Libraries and Dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
X=dataset.iloc[:,[3,4]].values
```

**Using Dendrograms to find Optimal Clusters**

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

**Training Hierarchical Clustering Model**

```python
from sklearn.cluster import AgglomerativeClustering
acobj=AgglomerativeClustering(n_clusters=3,linkage='ward')
y_pred=acobj.fit_predict(X)
```

**Visualizing Hierarchical Clustering Model**

```python
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1],s=50,c='red',label='c0')
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1],s=50,c='blue',label='c1')
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1],s=50,c='green',label='c2')
#plt.scatter(X[y_pred == 3,0],X[y_pred == 3,1],s=50,c='yellow',label='c3')
#plt.scatter(X[y_pred == 4,0],X[y_pred == 4,1],s=50,c='orange',label='c4')
plt.title('Clusters of Customers')
plt.xlabel('Annual Income(in k$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```
