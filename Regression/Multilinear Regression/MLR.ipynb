**Imporing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
dataset=pd.read_csv('Data.csv')
mof=dataset.iloc[:, :-1].values
dv=dataset.iloc[:,-1].values
```

```python
print("Matrix of features:\n",mof,"\n")
```

```python
print("Dependent variables:\n",dv)
```

**Splitting dataset into training and test set**

```python
from sklearn.model_selection import train_test_split
mof_train,mof_test,dv_train,dv_test=train_test_split(mof,dv,test_size=0.2,random_state=0)
```

```python
print("Training set:\n")
print("Matrix of feature:\n",mof_train,"\n")
print("Dependent Variables:\n",dv_train)
```

```python
print("Test set:\n")
print("Matrix of feature:\n",mof_test,"\n")
print("Dependent Variables:\n",dv_test)
```

**Training Model**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(mof_train,dv_train)
```

**Testing model**

```python
dv_prediction=lr.predict(mof_test)
np.set_printoptions(precision=2)
print(np.concatenate((dv_prediction.reshape(len(dv_prediction),1),dv_test.reshape(len(dv_test),1)),1))
```

**Evaluating Model**

```python
from sklearn.metrics import r2_score
r2_score(dv_test,dv_prediction)
```
