**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
dataset=pd.read_csv('Data.csv')
idv=dataset.iloc[:, :-1].values
dv=dataset.iloc[:,-1].values
```

**Spliting Dataset**

```python
from sklearn.model_selection import train_test_split
idv_train,idv_test,dv_train,dv_test=train_test_split(idv,dv,test_size=0.2,random_state=0)
```

**Training RFR model**

```python
from sklearn.ensemble import RandomForestRegressor
rfrobj=RandomForestRegressor(n_estimators=10,random_state=0)
rfrobj.fit(idv_train,dv_train)
```

**Testing RFR model**

```python
dv_prediction=rfrobj.predict(idv_test)
np.set_printoptions(precision=2)
print(np.concatenate((dv_prediction.reshape(len(dv_prediction),1),dv_test.reshape(len(dv_test),1)),1))
```

**Evaluating Model**

```python
from sklearn.metrics import r2_score
r2_score(dv_test,dv_prediction)
```
