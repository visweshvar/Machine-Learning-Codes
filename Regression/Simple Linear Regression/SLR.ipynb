**Importing Libraries**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
dataset = pd.read_csv('Salary_Data.csv')
mof = dataset.iloc[:, :-1].values
dv = dataset.iloc[:, -1].values
```

```python
print("Matrix of features:\n",mof,end="\n")
```

```python
print("Dependent Variable:\n",dv,end="\n")
```

**Splitting dataset into training and test set**

```python
from sklearn.model_selection import train_test_split
mof_train,mof_test,dv_train,dv_test = train_test_split(mof,dv,test_size=0.2, random_state=0)
```

```python
print("Matrix of features of training set:",mof_train,end="\n")
```

```python
print("Dependent variable of training set:",mof_test,end="\n")
```

```python
print("Matrix of features of test set:",dv_train,end="\n")
```

```python
print("Dependent variable of test set:",dv_test,end="\n")
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
mof_train=sc.fit_transform(mof_train)
mof_test=sc.fit_transform(mof_test)
```

```python
print(mof_train)
```

```python
print(mof_test)
```

**Training the model**

```python
from sklearn.linear_model import LinearRegression
lrobj=LinearRegression()
lrobj.fit(mof_train,dv_train)
```

Testing the model

```python
dv_prediction=lrobj.predict(mof_test)
```

 **Plotting Training set**

```python
plt.scatter(mof_train,dv_train,color="blue")
plt.plot(mof_train,lrobj.predict(mof_train),color="red")
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

Plotting Test set

```python
plt.scatter(mof_test,dv_test,color="blue")
plt.plot(mof_test,lrobj.predict(mof_test),color="red")
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```
