**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
dataset=pd.read_csv('Data.csv')
X=dataset.iloc[:, :-1].values
Y=dataset.iloc[:,-1].values
Y=Y.reshape(len(Y),1)
```

**Splitting Dataset**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sX=StandardScaler()
sY=StandardScaler()
X_train=sX.fit_transform(X_train)
Y_train=sY.fit_transform(Y_train)
```

**Training SVR Model**

```python
from sklearn.svm import SVR
svrobj=SVR(kernel='rbf')
svrobj.fit(X_train,Y_train)
```

**Predicting new results**

```python
y_prediction=sY.inverse_transform(svrobj.predict(sX.transform(X_test)).reshape(-1,1))
np.set_printoptions(precision=2)
print(np.concatenate((y_prediction.reshape(len(y_prediction),1),Y_test.reshape(len(Y_test),1)),1))
```

**Evaluating Model**

```python
from sklearn.metrics import r2_score
r2_score(Y_test,y_prediction)
```
