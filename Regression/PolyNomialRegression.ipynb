**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
dataset=pd.read_csv('Data.csv')
mof=dataset.iloc[:, :-1].values
dv=dataset.iloc[:,-1].values
```

**Splitting Dataset**

```python
from sklearn.model_selection import train_test_split
mof_train,mof_test,dv_train,dv_test=train_test_split(mof,dv,test_size=0.2,random_state=0)
```

**Training Polynomial Regression Model**

```python
 from sklearn.linear_model import LinearRegression
 from sklearn.preprocessing import PolynomialFeatures
 pfobj=PolynomialFeatures(degree=4)
 mof_poly =  pfobj.fit_transform(mof_train)
 lrobj1=LinearRegression()
 lrobj1.fit(mof_poly,dv_train)
```

**Testing Polynomial Regression Model**

```python
dv_prediction=lrobj1.predict(pfobj.transform(mof_test))
np.set_printoptions(precision=2)
print(np.concatenate((dv_prediction.reshape(len(dv_prediction),1),dv_test.reshape(len(dv_test),1)),1))
```

**Evaluating Model**

```python
from sklearn.metrics import r2_score
r2_score(dv_test,dv_prediction)
```
