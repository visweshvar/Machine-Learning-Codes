**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
dataset=pd.read_csv('Data.csv')
idv=dataset.iloc[:, :-1].values
dv=dataset.iloc[:,-1].values
```

**Splitting Dataset**

```python
from sklearn.model_selection import train_test_split
idv_train,idv_test,dv_train,dv_test=train_test_split(idv,dv,test_size=0.2,random_state=0)
```

**Trianining DT Model**

```python
from sklearn.tree import DecisionTreeRegressor
dtrobj=DecisionTreeRegressor(random_state=0)
dtrobj.fit(idv_train,dv_train)
```

**Testing the DT Model**

```python
dv_prediction=dtrobj.predict(idv_test)
np.set_printoptions(precision=2)
print(np.concatenate((dv_prediction.reshape(len(dv_prediction),1),dv_test.reshape(len(dv_test),1)),1))
```

**Evaluating Model**

```python
from sklearn.metrics import r2_score
r2_score(dv_test,dv_prediction)
```
